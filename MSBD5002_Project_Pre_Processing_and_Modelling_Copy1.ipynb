```python
#convert all numbers to words

import inflect

def number_to_word(number_str):
    
    p = inflect.engine()
    
    if   (len(number_str) <= 2):
        number = p.number_to_words(int(number_str))
        if(number == "zero"):
            return "ou"
        else: 
            return number.replace("-"," ")
    
    elif (len(number_str) == 4):
        digit_1 = int( number_str[:2] )
        digit_2 = int( number_str[2:] )
        number = p.number_to_words(digit_1) + " " + p.number_to_words(digit_2)
        return number.replace("-", " ")
        
    else:
        return " ".join( [p.number_to_words(int(digit)) for digit in number_str] )

print( number_to_word("55"), number_to_word("911"))
```

```python
def get_number_mapping( text ):
    words = text.split(" ")
    numbers_str = filter( lambda word: word.isnumeric(), words )
    
    mapping = {}
    for number in set(numbers_str):
        mapping[number] = number_to_word(number)
    return mapping

encoding = get_number_mapping(expanded)
encoding
```

# Play About - Building a character level model (LSTM)

```python
#Sources: https://towardsdatascience.com/generating-drake-rap-lyrics-using-language-models-and-lstms-8725d71b1b12
#         https://github.com/nikolaevra/drake-lyric-generator
```

```python
import numpy as np
```

```python
# create a dictionary mapping chracter-to-index
char_indices = dict((c, i) for i, c in enumerate(chars))
```

```python
# create a dictionary mapping index-to-chracter
indices_char = dict((i, c) for i, c in enumerate(chars))
```

```python
# cut the text into sequences
maxlen = 20
step = 1 # step size at every iteration
sentences = [] # list of sequences
next_chars = [] # list of next chracters that our model should predict
```

```python
def create_sequences(text, sequence_length, step):
    sequences = []
    next_chars = []
    for i in range(0, len(text) - sequence_length, step):
        sequences.append(text[i: i + sequence_length])
        next_chars.append(text[i + sequence_length])
    return sequences, next_chars
```

```python
def build_model(sequence_length, chars):
    model = Sequential()
    model.add(LSTM(128, input_shape=(sequence_length, len(chars))))
    model.add(Dense(len(chars)))
    model.add(Activation('softmax'))

    optimizer = RMSprop(lr=0.01)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer)
    return model
```

```python
def sample(preds, temperature=1.0):

    if temperature == 0:
        temperature = 1

    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    probas = np.random.multinomial(1, preds, 1)
    return np.argmax(probas)
```

# Building a Markov Model - Character Level Approach

```python
import string, random, sys

class MarkovModel:

    # create Markov model of order k for specified text
    def markovModel(self, text, k):
        # make circular
        start = text[0:k]
        text = " ".join([text, start])
        kgrams = {}
        for i in range(0, len(text)-k, 1):
            key = text[i:i+k] # kgram
            char = text[i+k] # next character
            # put in kgram table
            if key not in kgrams:
                kgrams[key] = {}
            if char not in kgrams[key]:
                kgrams[key][char] = 1
            else:
                kgrams[key][char] += 1
        return kgrams
    
    # get next character in model by weighted probability
    def nextCharacter(self, kgrams, key):
        choice = []
        for letter in kgrams[key]:
            for count in range(0, kgrams[key][letter], 1):
                choice.append(letter)
        return random.choice(choice)   
    
    # generate text character by character
    def textGenerator(self, text, k, characters):
        kgrams = self.markovModel(text, k)
        rand_idx = random.randint(0,len(text)-k-1)
        starting_kgram = text[rand_idx:rand_idx+k]
        lyrics = [starting_kgram]
        
        print(lyrics[0])
        for i in range(0, characters, 1):
            # print character
            char = self.nextCharacter(kgrams, starting_kgram)
            lyrics.append(char)
            print(char,end='')
            # update kgram
            starting_kgram = starting_kgram[1:] # remove first character
            starting_kgram = "".join([starting_kgram,char])# add new character

        return lyrics   
    
    # generate string of lyrics
    def stringLyrics(self, lyrics):
        return "".join(lyrics)
```

# Another Markov Model - Word Level Approach (using pairs of words)

```python
#Source: https://towardsdatascience.com/simulating-text-with-markov-chains-in-python-1a27e6d13fc6
#        https://gist.github.com/bpshaver/840d53222b72d2612ddf6702ef020274#file-markov_text_sim-py
```

```python
# split the lyrics to get single words
corpus = cleaned_lyrics.split()
```

```python
#define a function to give us all the pairs of words in the lyrics
def make_pairs(corpus):
    for i in range(len(corpus)-1):
        yield (corpus[i], corpus[i+1])
        
pairs = make_pairs(corpus)
```

```python
#we instantiate an empty dictionary, and fill it words from our pairs. If the first word of the pair is already a key 
#in the dictionary, simply append the next word to the list of words that follow that word. Otherwise, initialize a new 
#entry in the dictionary with the key equal to the first word and the value a list of length one

word_dict = {}
for word_1, word_2 in pairs:
    if word_1 in word_dict.keys():
        word_dict[word_1].append(word_2)
    else:
        word_dict[word_1] = [word_2]
```

```python
#we pick some random word to kick off the chain, and choose the number of words we want to simulate

first_word = np.random.choice(corpus)
chain = [first_word]
n_words = 30
```

```python
#After the first word, every word in the chain is sampled randomly from the list of words which have followed that word 
#in the lyrics:

for i in range(n_words):
    chain.append(np.random.choice(word_dict[chain[-1]]))
```

```python
#SOURCE: https://www.kdnuggets.com/2019/11/markov-chains-train-text-generation.html
#        https://github.com/StrikingLoo/ASOIAF-Markov
```

```python
corpus_words = cleaned_lyrics.split(' ')
corpus_words= [word for word in corpus_words if word != '']
corpus_words
len(corpus_words) # 2185920
```

```python
distinct_words = list(set(corpus_words))
word_idx_dict = {word: i for i, word in enumerate(distinct_words)}
distinct_words_count = len(list(set(corpus_words)))
distinct_words_count
```

```python
k = 2 # adjustable
sets_of_k_words = [ ' '.join(corpus_words[i:i+k]) for i, _ in enumerate(corpus_words[:-k]) ]

from scipy.sparse import dok_matrix

sets_count = len(list(set(sets_of_k_words)))
next_after_k_words_matrix = dok_matrix((sets_count, len(distinct_words)))

distinct_sets_of_k_words = list(set(sets_of_k_words))
k_words_idx_dict = {word: i for i, word in enumerate(distinct_sets_of_k_words)}

for i, word in enumerate(sets_of_k_words[:-k]):

    word_sequence_idx = k_words_idx_dict[word]
    next_word_idx = word_idx_dict[corpus_words[i+k]]
    next_after_k_words_matrix[word_sequence_idx, next_word_idx] +=1
```

```python
def sample_next_word_after_sequence(word_sequence, alpha = 0):
    next_word_vector = next_after_k_words_matrix[k_words_idx_dict[word_sequence]] + alpha
    likelihoods = next_word_vector/next_word_vector.sum()
    
    return weighted_choice(distinct_words, likelihoods.toarray())

import random
from random import random 

def weighted_choice(objects, weights):
    """ returns randomly an element from the sequence of 'objects', 
        the likelihood of the objects is weighted according 
        to the sequence of 'weights', i.e. percentages."""

    weights = np.array(weights, dtype=np.float64)
    sum_of_weights = weights.sum()
    # standardization:
    np.multiply(weights, 1 / sum_of_weights, weights)
    weights = weights.cumsum()
    x = random()
    for i in range(len(weights)):
        if x < weights[i]:
            return objects[i]


def stochastic_chain(seed, chain_length=15, seed_length=2):
    current_words = seed.split(' ')
    if len(current_words) != seed_length:
        raise ValueError(f'wrong number of words, expected {seed_length}')
    sentence = seed

    for _ in range(chain_length):
        sentence+=' '
        next_word = sample_next_word_after_sequence(' '.join(current_words))
        sentence+=next_word
        current_words = current_words[1:]+[next_word]
    return sentence
# example use    
stochastic_chain('the world')
```
